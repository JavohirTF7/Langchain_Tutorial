# **Using Gemini API**

### API kalitlarini xavfsiz kiritish


```python
from getpass import getpass
from google import genai

api_key = getpass("Gemini API kalitingizni kiriting: ")

client = genai.Client(api_key=api_key)
```

### Modelni ishlatib ko'rish

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Su'niy intellekt qanday ishlaydi, 1 ta gap bilan tushuntir"
)

print(response.text)
```

## Chat rejimida ishlatish ("assistant" agent)

```python
chat = client.chats.create(model="gemini-2.5-flash")

response = chat.send_message("salom, sen nimalar qila olasan?")
print(response.text)
```
